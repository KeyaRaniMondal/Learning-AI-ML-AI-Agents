```python
import pandas as pd
import seaborn as sns #used for making statistical graph
import matplotlib.pyplot as plt #used for creating static, interactive, and animated visualizations in Python
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv(r"D:\dataset\test.csv")
dataset.head(3)
```

```python
plt.figure(figsize=(10,15))
sns.scatterplot(x="x",y="y",data=dataset)
plt.show()
```

```python
dataset.isnull().sum()
```

```python
#data should always be multidimensional(means 2 dimensional)
ind=dataset["x"]
ind.ndim #for checking dimensions
```

```python
ind=dataset[["x"]] #for creating 1 dimension to 2d
ind.ndim
```

```python
x=dataset[["x"]] #for creating 1 dimension to 2d
y=dataset["y"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)#0.2 means 20% of data , 42 is a standard number
# to retrain for best accuracy we need to change the value of random state from 1~ and at which point it gives the best accuracy we will note it 
```

```python
from sklearn.linear_model import LinearRegression #it takes the equation y=mx+c
```

```python
lr=LinearRegression() #it is the object
lr.fit(x_train,y_train) #it gives the best value of m and c to prefectly draw straight line
```

```python
lr.coef_ #returns coefficient means the value of m
```

```python
lr.intercept_ #returns the value of intercept means the value of c
```

```python
lr.score(x_test,y_test)*100 # if it is predicting values correctly or not ,wwe need to check the accuracy score
```

```python
lr.predict([[77]]) #it helps to predict
```

```python
y_pred=lr.predict(x)
```

```python
plt.figure(figsize=(8,12))
sns.scatterplot(x="x",y="y",data=dataset)
plt.plot(dataset["x"],y_pred,c="red") # to add the straight line over the graph
plt.legend(["org data","predict line"]) #for showing label
# plot.savefig("predict.jpg") for getting the graph on pc
plt.show()
```
